Model Evaluation

```python
# akurasi
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

diabetes_data = load_breast_cancer()

df_breast_cancer = pd.DataFrame(data=diabetes_data.data,
                                columns=diabetes_data.feature_names)

df_breast_cancer.head()

```

```python
diabetes_data
```

```python
df_breast_cancer.info()
```

```python
# splitting data
X = diabetes_data.data
y = diabetes_data.target

# splittingnya di sini
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2,
                                                    random_state = 24)

print(f"Train Data: {X_train.shape}, {y_train.shape}")
print(f"Test Data: {X_test.shape}, {y_test.shape}")
```

```python
# Bangun model atau classified, inisiasis object
classifier = RandomForestClassifier(random_state=18)
classifier.fit(X_train, y_train)
```

```python
# data testing
predictions = classifier.predict(X_test)
```

```python
# menampilkan conclusion matrix
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

print(confusion_matrix(y_test, predictions))

plot_confusion_matrix(classifier, X_test, y_test)
plt.show()
```

```python
# Hitung akurasi
from sklearn.metrics import accuracy_score

print(f"Accuracy of the classifier is : {accuracy_score(y_test, predictions)}")
```

```python
TP = 69
TN = 39
FP = 2
FN = 4

acc = (TP+TN)/(TP+TN+FP+FN)
print(acc)
```

Precision

```python
# Hitung presisi
from sklearn.metrics import precision_score

print(f"Precision of the classifier is : {precision_score(y_test, predictions)}")
```

```python
TP = 69
TN = 39
FP = 2
FN = 4

precision = (TP)/(TP+FP)
print(precision)
```

Recall/Sensitivity/True Positive Rate

```python
# Hitung recall
from sklearn.metrics import recall_score

print(f"Recall Score of the classifier is : {recall_score(y_test, predictions)}")
```

```python
TP = 69
TN = 39
FP = 2
FN = 4
# True positive rate
recall = (TP)/(TP+FN)
print(recall)
```

Specificity

```python
# Hitung specify
from imblearn.metrics import specificity_score

print(f"Specificity Score of the classifier is : {specificity_score(y_test, predictions)}")
```

```python
TP = 69
TN = 39
FP = 2
FN = 4

specificity = (TN)/(TN+FP)
print(specificity)
```

```python
# F1 score
from sklearn.metrics import f1_score

print(f"F1 Score of the classifier is : {f1_score(y_test, predictions)}")
```

```python
f1score = 2*precision*recall/(precision+recall)
print(f1score)
```

Classification report

```python
# F1 score
from sklearn.metrics import classification_report

print({classification_report(y_test, predictions, zero_division=0)})
```

ROC

```python
from sklearn.metrics import roc_curve, auc
class_probabilities = classifier.predict_proba(X_test)
preds = class_probabilities[:, 1]
fpr, tpr, threshold = roc_curve(y_test, preds)
roc_auc = auc(fpr, tpr)

print(f"AUC for our classifier is: (roc_auc")

plt.title('Receiver Operating Charactheristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
