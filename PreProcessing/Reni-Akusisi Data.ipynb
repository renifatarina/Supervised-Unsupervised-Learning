```python
#import tweepy
import tweepy
```

```python
#masukkan access token
#token sesuai akun masing2
api_key="xxx"
api_secret_key="xxx"

access_token="xxx"
access_token_secret="xxx"
```

```python
auth=tweepy.OAuthHandler(api_key,api_secret_key)
auth.set_access_token(access_token,access_token_secret)
api=tweepy.API(auth)
```

```python
posts = api.user_timeline(screen_name="jokowi", count = 100, lang ="id", tweet_mode="extended")
```

```python
print("Berikut ini adalah 5 twit terakhir dari akun @jokowi:\n")
i=1
for tweet in posts[:5]:
    print(str(i) +') '+ tweet.full_text + '\n')
    i= i+1
```

```python
# membuat dataframe dengan kolom tweets
import pandas as pd
df = pd.DataFrame([ tweet.full_text.lower() for tweet in posts], columns = ['Tweets'])
```
